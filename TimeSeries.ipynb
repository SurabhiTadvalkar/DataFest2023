```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('data/questions.csv')
df = df.rename(columns = {"AskedOnUtc" : "datetime"})
df["datetime"] = pd.to_datetime(df["datetime"], format = '%Y-%m-%d %H:%M:%S')
df["date"] = df["datetime"].dt.date
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df = df[df["date"] > datetime.date(2016, 1, 1)]

df["date"] = df["date"].astype(str)
df[["year", "month+day"]] = df["date"].str.split('-', n=1, expand=True)
df["month+day"] = ("2000-" + df["month+day"])
df["time"] = df["datetime"].dt.time
df
```

```python
CA = df[df["StateAbbr"] == 'CA']
CA = CA.rename(columns = {"AskedOnUtc" : "datetime"})
CA["datetime"] = pd.to_datetime(CA["datetime"], format = '%Y-%m-%d %H:%M:%S')
CA["date"] = CA["datetime"].dt.date
CA["time"] = CA["datetime"].dt.time
CA
```

```python
ct = pd.crosstab(df["month+day"], df["Category"])
ct["All"] = ct.sum(axis = 1)
ct = ct/6
ct
```
